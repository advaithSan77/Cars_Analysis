```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('USA_cars_datasets.csv')
```

```python
df
```

```python
del df['Unnamed: 0']
del df['condition']
del df['lot']
```

```python
df.head(10)
```

```python
df.isna().sum()
```

#No. of cars purchased of each brand

```python
brand_nos = pd.DataFrame(df.groupby('brand', as_index=False, sort=True)['vin'].count())
brand_nos.sort_values(by=['vin'], ascending=False, inplace=True)
brand_nos.rename(columns={'vin':'No of cars'}, inplace=True)
brand_nos
```

```python
graph = plt.figure(figsize=(25, 10))
plt.bar(brand_nos['brand'], brand_nos['No of cars'], width=0.5)
plt.xlabel('Brand')
plt.ylabel('Number of cars')
plt.title('No. of cars purchased of each brand')
plt.xticks(rotation=70)
plt.show
```

#No. of cars purchased each year

```python
y_nos = pd.DataFrame(df.groupby(['year'], as_index=False, sort=True)['vin'].count())
#y_nos.sort_values(by='vin', ascending=False, inplace=True)
y_nos.rename(columns={'vin':'No of cars'}, inplace=True)
y_nos
```

```python
graph = plt.figure(figsize=(25, 10))
plt.plot(y_nos['year'], y_nos['No of cars'], color='black')
plt.xlabel('Year')
plt.ylabel('Number of cars')
plt.title('No. of cars purchased each year')
plt.xticks()
plt.show
```

#No. of cars of each color

```python
brand_y_nos = pd.DataFrame(df.groupby(['color'], as_index=False, sort=True)['vin'].count())
brand_y_nos.sort_values(by='vin', ascending=False, inplace=True)
brand_y_nos.rename(columns={'vin':'No of cars'}, inplace=True)
brand_y_nos.head(20)
```

```python
graph = plt.figure(figsize=(25, 10))
plt.bar(brand_y_nos['color'], brand_y_nos['vin'], color='blue', width=0.6)
plt.xlabel('Color')
plt.ylabel('Number of cars')
plt.title('No. of cars of each color')
plt.xticks(rotation=80)
plt.show
```

#No. of cars of each model

```python
mod_nos = pd.DataFrame(df.groupby(['model'], as_index=False, sort=True)['vin'].count())
mod_nos.sort_values(by='vin', ascending=False, inplace=True)
mod_nos.rename(columns={'vin':'No of cars'}, inplace=True)
mod_nos.head(20)
```

```python
graph = plt.figure(figsize=(28, 10))
plt.bar(mod_nos['model'], mod_nos['No of cars'], color='red', width=0.6)
plt.xlabel('Model')
plt.ylabel('Number of cars')
plt.title('No. of cars of each model')
plt.xticks(rotation=80)
plt.show
```

#Maximum car price each year

```python
year_max = pd.DataFrame(df.groupby(['year'], as_index=False, sort=True)['price'].max())
year_max
```

```python
graph = plt.figure(figsize=(15, 8))
plt.plot(year_max['year'], year_max['price'], color='black')
plt.xlabel('Year')
plt.ylabel('Maximum car price')
plt.title('Maximum car price each year')
plt.show
```

#Average car price each year

```python
year_mean = pd.DataFrame(df.groupby(['year'], as_index=False, sort=True)['price'].mean())
year_mean
```

```python
graph = plt.figure(figsize=(15, 8))
plt.plot(year_mean['year'], year_mean['price'], color='black')
plt.xlabel('Year')
plt.ylabel('Average car price')
plt.title('Average car price each year')
plt.show
```

#No of cars in each state

```python
state_nos = pd.DataFrame(df.groupby('state', as_index=False, sort=True)['vin'].count())
state_nos.sort_values(by=['vin'], ascending=False, inplace=True)
state_nos.rename(columns={'vin':'No of cars'}, inplace=True)
state_nos
```

```python
graph = plt.figure(figsize=(15, 8))
plt.bar(state_nos['state'], state_nos['No of cars'], color='black', width=0.5)
plt.xlabel('State')
plt.ylabel('No of cars ')
plt.title('No of cars in each state')
plt.xticks(rotation=70)
plt.show
```

#State with maximum average price

```python
state_max = pd.DataFrame(df.groupby('state', as_index=False, sort=True)['price'].mean())
state_max.sort_values(by=['price'], ascending=False, inplace=True)
state_max
```

```python
graph = plt.figure(figsize=(15, 8))
plt.bar(state_max['state'], state_max['price'], color='purple', width=0.5)
plt.xlabel('State')
plt.ylabel('Average price')
plt.title('Average prices in states')
plt.xticks(rotation=70)
plt.show
```
